```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
image = cv2.imread('data/Input/romboide-color.jpg', cv2.IMREAD_GRAYSCALE)

# Mostrar la imagen original
plt.figure(figsize=(12, 6))
plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

# Binarización inversa con Otsu
_, binary_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

plt.subplot(2, 3, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Binaria INV (Otsu)')
plt.axis('off')

# Kernel para operaciones morfológicas
kernel = np.ones((3, 3), np.uint8)

# Apertura + cierre
opening = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)

# Detectar contornos
contours, _ = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Crear máscara vacía y rellenar el contorno principal
mask = np.zeros_like(closing)
cv2.drawContours(mask, contours, -1, 255, thickness=cv2.FILLED)

plt.subplot(2, 3, 3)
plt.imshow(mask, cmap='gray')
plt.title('Figura Rellena')
plt.axis('off')

# Dibujar los contornos en la imagen original
contour_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

plt.subplot(2, 3, 4)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title('Contornos detectados')
plt.axis('off')

# Calcular área y perímetro sobre la figura rellenada
areas = [cv2.contourArea(c) for c in contours]
perimeters = [cv2.arcLength(c, True) for c in contours]

print("Áreas detectadas:", areas)
print("Perímetros detectados:", perimeters)

# Mostrar figura rellenada como resultado final
plt.subplot(2, 3, 5)
plt.imshow(mask, cmap='gray')
plt.title('Resultados finales')
plt.axis('off')

# Gráfico de características
plt.subplot(2, 3, 6)
plt.plot(areas, label="Áreas", color="blue", marker='o')
plt.plot(perimeters, label="Perímetros", color="red", marker='x')
plt.legend()
plt.title('Características de los objetos')
plt.xlabel('Objeto indexado')
plt.ylabel('Valor')

plt.tight_layout()
plt.show()
```
