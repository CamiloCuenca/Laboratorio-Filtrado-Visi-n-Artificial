1. Cambiar de RGB a Escala de Grises

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
img = cv2.imread('data/Input/romboide-sucio.png')

# Convertir la imagen de BGR a escala de grises
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Mostrar la imagen en escala de grises
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()

# Guardar la imagen en escala de grises
cv2.imwrite("data/generated/Op/romboide-gris.jpg", gray_img)
```

2. Convertir a Blanco y Negro (Binaria)

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
img = cv2.imread('data/Input/romboide-sucio.png')

# Convertir a escala de grises antes de binarizar
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Convertir la imagen a blanco y negro (binaria)
_, binary_img = cv2.threshold(gray_img, 127, 250, cv2.THRESH_BINARY)

# Mostrar la imagen en blanco y negro
plt.imshow(binary_img, cmap='gray')  
plt.axis('off')
plt.show()

# Guardar la imagen en escala de grises
cv2.imwrite("data/generated/Op/romboide-binarizado.jpg", binary_img)
```

3. Cambiar de Tamaño (Reducción o Amplificación)

```python
import cv2
import matplotlib.pyplot as plt

# Cambiar el tamaño de la imagen (Reducción y Amplificación)
resized_img = cv2.resize(gray_img, (400, 400))  # Cambiar el tamaño a 400x400 píxeles

# Mostrar la imagen redimensionada
plt.imshow(resized_img)
plt.axis('off')  # Ocultar los ejes
plt.show()


# Guardar la imagen en escala de grises
cv2.imwrite("data/generated/Op/romboide-redimensionada.jpg", resized_img)
```

4. Operaciones Geométricas: Rotación

```python
import cv2
import numpy as np

# Obtener dimensiones de la imagen
(h, w ) = img.shape[:2]

# Encontrar el centro de la imagen
centro = (w // 2, h // 2)

# Matris de rotacion (50 grados)
M = cv2.getRotationMatrix2D(centro, 50, 1.0)

# Rotar la imagen
rotada = cv2.warpAffine(img, M, (w, h))

# Mostrar la imagen rotada
plt.imshow(rotada)
plt.axis('off')  # Ocultar los ejes
plt.show()

# Guardar la imagen rotada
cv2.imwrite("data/generated/Op/romboide-rotado.jpg", rotada)
```

5. Operaciones Aritméticas: Suma, Resta, Multiplicación y División

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
# Cargar dos imágenes (deben tener el mismo tamaño)
img1 = cv2.imread('data/Input/romboide-sucio.png')
img2 = cv2.imread('data/Input/romboide-color.jpg')

# Suma de imágenes
sum_img = cv2.add(img1, img2)

# Resta de imágenes
diff_img = cv2.subtract(img1, img2)

# Multiplicación de imágenes
mult_img = cv2.multiply(img1, img2)

# División de imágenes
div_img = cv2.divide(img1, img2)

# Mostrar las imágenes resultantes
fig, axes = plt.subplots(2,2 , figsize=(15, 5))

axes[0][0].imshow(sum_img, cmap='gray')
axes[0][0].set_title("Suma")
axes[0][0].axis('off')

axes[0][1].imshow(diff_img, cmap='gray')
axes[0][1].set_title("Resta")
axes[0][1].axis('off')

axes[1][0].imshow(mult_img, cmap='gray')
axes[1][0].set_title("Multiplicación")
axes[1][0].axis('off')

axes[1][1].imshow(div_img, cmap='gray')
axes[1][1].set_title("División")
axes[1][1].axis('off')

plt.show()


# Guardar las imágenes
cv2.imwrite('data/generated/Op/romboide-suma.jpg', sum_img)
cv2.imwrite('data/generated/Op/romboide-resta.jpg', diff_img)
cv2.imwrite('data/generated/Op/romboide-multiplicacion.jpg', mult_img)
cv2.imwrite('data/generated/Op/romboide-division.jpg', div_img)
```

6. Operaciones Lógicas: AND, OR, NOT

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Cargar dos imágenes binarizadas 
img1 = cv2.imread('data/Input/romboide-sucio.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('data/Input/romboide-color.jpg', cv2.IMREAD_GRAYSCALE)

# Convertir a binario 
_, img1_bin = cv2.threshold(img1, 128, 255, cv2.THRESH_BINARY)
_, img2_bin = cv2.threshold(img2, 128, 255, cv2.THRESH_BINARY)

# Operación AND
and_img = cv2.bitwise_and(img1_bin, img2_bin)

# Operación OR
or_img = cv2.bitwise_or(img1_bin, img2_bin)

# Operación NOT
not_img1 = cv2.bitwise_not(img1_bin)
not_img2 = cv2.bitwise_not(img2_bin)

# Mostrar las imágenes resultantes

# Mostrar las imágenes utilizando matplotlib
fig, axes = plt.subplots(2,2 , figsize=(15, 5))

axes[0][0].imshow(and_img, cmap='gray')
axes[0][0].set_title("AND")
axes[0][0].axis('off')

axes[0][1].imshow(or_img, cmap='gray')
axes[0][1].set_title("OR")
axes[0][1].axis('off')

axes[1][0].imshow(not_img1, cmap='gray')
axes[1][0].set_title("NOT IMAGEN UNO")
axes[1][0].axis('off')

axes[1][1].imshow(not_img2, cmap='gray')
axes[1][1].set_title("NOT IMAGEN DOS")
axes[1][1].axis('off')

plt.show()


# Guardar las imágenes
cv2.imwrite('data/generated/Op/romboide-and.jpg', and_img)
cv2.imwrite('data/generated/Op/romboide-or.jpg', or_img)
cv2.imwrite('data/generated/Op/romboide-not1.jpg', not_img1)
cv2.imwrite('data/generated/Op/romboide-not2.jpg', not_img2)
```

7. Interpolación (Reducción y Amplificación)

```python
import cv2

# Cargar la imagen
img = cv2.imread('data/Input/romboide-sucio.png')

# Cambiar tamaño usando diferentes métodos de interpolación
resized_bilinear = cv2.resize(img, (400, 400), interpolation=cv2.INTER_LINEAR)  # Bilineal
resized_nearest = cv2.resize(img, (400, 400), interpolation=cv2.INTER_NEAREST)  # Vecino más cercano

# Mostrar las imágenes utilizando matplotlib
fig, axes = plt.subplots(2, 1, figsize=(15, 5))

# Escala de grises
axes[0].imshow(resized_bilinear, cmap='gray')
axes[0].set_title("Interpolación Bilineal")
axes[0].axis('off')

# Blanco y negro (binaria)
axes[1].imshow(resized_nearest, cmap='gray')
axes[1].set_title("Interpolación Vecino más cercano")
axes[1].axis('off')

plt.show()


cv2.imwrite('data/generated/Op/romboide-interpolacion-bilinear.jpg', resized_bilinear)
cv2.imwrite('data/generated/Op/romboide-interpolacion-vecino.jpg', resized_nearest)
```
