1. Cambiar de RGB a Escala de Grises

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
img = cv2.imread('data/Input/romboide-color.jpg')

# Convertir la imagen de BGR a escala de grises
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Mostrar la imagen en escala de grises
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()

# Guardar la imagen en escala de grises
cv2.imwrite("data/generated/romboide-gris.jpg", gray_img)
```

2. Convertir a Blanco y Negro (Binaria)

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
img = cv2.imread('data/Input/romboide-color.jpg')

# Convertir a escala de grises antes de binarizar
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Convertir la imagen a blanco y negro (binaria)
_, binary_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

# Mostrar la imagen en blanco y negro
plt.imshow(binary_img, cmap='gray')  
plt.axis('off')
plt.show()

# Guardar la imagen en escala de grises
cv2.imwrite("data/generated/romboide-binarizado.jpg", binary_img)
```
